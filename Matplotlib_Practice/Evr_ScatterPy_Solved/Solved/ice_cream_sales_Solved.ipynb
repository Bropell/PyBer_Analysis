```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
temp = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2]
sales = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]
```

```python
# Tell matplotlib to create a scatter plot based upon the above data

plt.scatter(temp, sales, color="skyblue", marker='o', edgecolor="black")

# Create labels for the x and y axes

plt.xlabel("Temperature ($^\circ$C)")
plt.ylabel("Sales ($USD)")


# Create a title

plt.title("Ice Cream Sales at Different Temperatures")


# Create limits for x and y axes

plt.xlim(min(temp) - 5, max(temp) + 5)
plt.ylim(min(sales) - 50, max(sales) + 50)

plt.minorticks_on()

# Save an image of the chart and print to screen

plt.savefig('ice_cream_sales.png')
plt.show()
```
